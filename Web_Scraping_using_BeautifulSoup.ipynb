```python
# Importer les packages
import pandas as pd
from requests import get # récupérer le contenu html de la page
from bs4 import BeautifulSoup as bs # s'assurer d'avoir le contenu dans un objet bs bien sparsifié
```

```python
# url
url = 'https://sn.coinafrique.com/categorie/vetements-homme'
```

```python
#Recuperer le contnenu de la page
res= get(url)
#avoir le contenu dans un objet bs et bien sparsifier
soup = bs(res.content, 'html.parser')
#recuperer une ou des information sur la page en partant d'une balise
# find  __________________ permet de recuperer une information specifique en partant d'une
# find_all _________________permet de recuperer plussieur information en partant d'un_ale balise

```

```python
# recuperer les conteneurs
containers = soup.find_all("div","col s6 m4 l3") # renvoie une liste du code qui se trouve dans cette balise exemple des balise p etc
len(containers)
```

```python
# premier container
container = containers[0]
#type_habits
type_habits = container.find('p','ad__card-description').a.text.strip()

#price
price_text = container.find('p','ad__card-price').a.text
price_text = price_text.strip().replace('CFA', '').replace(' ', '')
price = float(price_text)

price

#adresse
adresse = container.find('p','ad__card-location').span.text.strip()
adresse

#url image
url_image = container.find('img','ad__card-img')['src']

print(type_habits , '---',price,'---',adresse,'---',url_image)
```

```python
# Importer les packages
import pandas as pd
from requests import get
from bs4 import BeautifulSoup as bs


def scraper_coinafrique(categorie, nom_colonne_type, nb_pages=30):
    """
    Scrape les données de CoinAfrique pour une catégorie donnée

    Args:
        categorie: nom de la catégorie dans l'URL (ex: 'vetements-homme')
        nom_colonne_type: nom de la colonne pour le type de produit (ex: 'type_habits')
        nb_pages: nombre de pages à scraper (par défaut 30)

    Returns:
        DataFrame avec les données scrapées
    """
    df_final = pd.DataFrame()

    for ind_page in range(1, nb_pages + 1):
        url = f'https://sn.coinafrique.com/categorie/{categorie}?page={ind_page}'
        res = get(url)

        # Parser le contenu HTML
        soup = bs(res.content, 'html.parser')
        containers = soup.find_all("div", "col s6 m4 l3")
        data = []

        for container in containers:
            try:
                # Type de produit
                type_produit = container.find('p', 'ad__card-description').a.text.strip()

                # Prix
                price_text = container.find('p', 'ad__card-price').a.text
                price_text = price_text.strip().replace('CFA', '').replace(' ', '')
                price = float(price_text)

                # Adresse
                adresse = container.find('p', 'ad__card-location').span.text.strip()

                # URL image
                url_image = container.find('img', 'ad__card-img')['src']

                # Créer le dictionnaire avec le nom de colonne dynamique
                dic = {
                    nom_colonne_type: type_produit,
                    'price': price,
                    'adresse': adresse,
                    'url_image': url_image
                }
                data.append(dic)
            except:
                pass

        # Créer et concaténer le DataFrame
        df_page = pd.DataFrame(data)
        df_final = pd.concat([df_final, df_page], axis=0).reset_index(drop=True)

    return df_final


# Utilisation de la fonction pour les 4 catégories
df_url1 = scraper_coinafrique('vetements-homme', 'type_habits')
df_url2 = scraper_coinafrique('chaussures-homme', 'type_chaussure')
df_url3 = scraper_coinafrique('vetements-enfants', 'type_habits')
df_url4 = scraper_coinafrique('chaussures-enfants', 'type_chaussures')
```
